```python
hairstyles = ["bouffant", "pixie", "dreadlocks", "crew", "bowl", "bob", "mohawk", "flattop"]
prices = [30, 25, 40, 20, 20, 35, 50, 35]
last_week = [2, 3, 5, 8, 4, 4, 6, 2]
```

```python
# 가격의 총합 계산
total_price = 0
for price in prices:
    total_price += price

print("가격의 총합: $" + str(total_price))
# 평균 가격 계산
average_price = total_price / len(prices)

# 평균 가격 출력
print("평균 가격: $" + str(average_price))
```

```python

new_prices = [price - 5 for price in prices]


print( new_prices)
```

```python
# 총 수익 계산을 위한 변수 초기화
total_revenue = 0

# 각 헤어스타일별로 발생한 수익 계산
for i in range(len(hairstyles)):
    total_revenue += prices[i] * last_week[i]

# 총 수익 출력
print("지난 주 총 수익: $" + str(total_revenue))

# 평균 일일 수익 계산
average_daily_revenue = total_revenue / 7

# 평균 일일 수익 출력
print("평균 일일 수익: $" + str(average_daily_revenue))
```

```python
# 30달러 미만의 헤어스타일을 모아서 리스트 생성
cuts_under_30 = [hairstyles[i] for i in range(len(new_prices)) if new_prices[i] < 30]

# 30달러 미만 헤어스타일 출력
print("30달러 미만의 헤어스타일:", cuts_under_30)
```
