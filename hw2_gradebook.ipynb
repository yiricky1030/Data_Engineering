```python
gradebook = [
    ["physics", 98],
    ["calculus", 97],
    ["poetry", 85],
    ["history", 88]
]
```

```python
print(gradebook)
```

```python
gradebook.append(["computer science", 100])
print(gradebook)
```

```python
gradebook.append(["visual arts", 93])
print(gradebook)
```

```python

visual_arts_index = None
for i, item in enumerate(gradebook):
    if item[0] == "visual arts":
        visual_arts_index = i
        break

if visual_arts_index is not None:
    gradebook[visual_arts_index][1] += 5


print(gradebook)
```

```python

poetry_index = None
for i, subject_grade in enumerate(gradebook):
    if subject_grade[0] == "poetry":
        poetry_index = i
        break

if poetry_index is not None:

    gradebook[poetry_index][1] = "Pass"


print(gradebook)
```

```python
last_semester_gradebook = [["politics", 80], ["latin", 96], ["dance", 97], ["architecture", 65]]
```

```python

full_gradebook = last_semester_gradebook + gradebook

print("Complete Gradebook:")
print(full_gradebook)
```
